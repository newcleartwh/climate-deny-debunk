```julia
using Plots
x = range(1997,2018, step=1)
#
y = [14.4, 14.5, 14.3, 14.3, 14.4, 14.5, 14.5, 14.4, 14.5, 14.5, 14.5, 14.4, 14.5, 14.6, 14.4, 14.5, 14.5, 14.6, 14.8, 14.8, 14.7, 14.6]
plot(x,y,label="@fmeeus1 cherry-picked data",smooth=true)

using DataFrames, GLM
data = DataFrame(X=x, Y=y)
ols = lm(@formula(Y ~ X), data)
predy = round.(predict(ols), digits=5)
scatter!(x,predy, seriestype = :line, label="Stijgende trend met Lineaire Regressie")
yaxis!("Temperatuur (in C°)")
xaxis!("Jaar (handgekozen door @fmeeus1)")

```
